Count null values in each columns

```python
null_val_counts = data.isnull().sum().to_frame(name="null_val_count")
null_val_counts = null_val_counts.reset_index()
null_val_counts = null_val_counts.rename(columns={"index": "column_names"})
null_val_counts["null_proportion(%)"] = null_val_counts["null_val_count"].apply(lambda x: x*100/len(data))
print(null_val_counts)
```

Lots of features have a large percentage of null values, simply drop those columns with null values of more than 40%
Meanwhile, drop the "id" column, which is nominal, and the "gcs_unable", which seems is to indicate whether the record has gcs value.

```python
columns_to_drop = null_val_counts[null_val_counts["null_proportion(%)"] >= 40].loc[:, "column_names"].values
columns_to_drop = np.append(columns_to_drop, ["id", "gcs_unable"])
print(columns_to_drop)
data = data.drop(columns_to_drop, axis=1)
```

See remain columns

```python
remain_columns = data.columns.values
print(remain_columns)
print("remain column nums: ",len(remain_columns))
```
